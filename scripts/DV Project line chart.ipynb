```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('owid-covid-data.csv')
print(df)
```

```python
df = df[df['continent'] == 'Asia'] 
df['year'] = pd.DatetimeIndex(df['date']).year
df['month'] = pd.DatetimeIndex(df['date']).month
df
```

```python
dff = df.groupby(by=["month","year","location"])['new_cases'].sum().sort_values(ascending=False)
print(dff)
```

```python
grouped_region_data = dff.reset_index()
# grouped_region_data['ratio'] = grouped_region_data['new_deaths']/grouped_region_data['new_vaccinations']
grouped_region_data.replace([np.inf, -np.inf], 0, inplace=True)
grouped_region_data.replace(np.nan, 0, inplace=True)
grouped_region_data.replace(to_replace="NULL",value=0)


grouped_region_data = grouped_region_data[grouped_region_data['year'] == 2020] 
print(grouped_region_data[grouped_region_data['year'] == 2020])
```
