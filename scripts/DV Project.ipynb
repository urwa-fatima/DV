```python
import pandas as pd
```

```python
df = pd.read_csv('owid-covid-data.csv')
print(df)
```

```python
df = df[df['continent'] == 'Asia'] 
df['year'] = pd.DatetimeIndex(df['date']).year
df
```

```python
dff = df.groupby(by=["year","location"])['new_deaths', 'new_vaccinations'].sum()
print(dff)
```

```python
grouped_region_data = dff.reset_index()
# grouped_region_data = grouped_region_data[grouped_region_data['year'] == 2020] 
grouped_region_data
```
