```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('owid-covid-data.csv')
print(df)
```

```python
df = df[df['continent'] == 'Asia'] 
```

```python
df['wave'] = 0
df.loc[(df['date'] > '2020-01-01') & (df['date'] < '2020-07-01'),"wave"] = 1
df.loc[(df['date'] > '2020-06-31') & (df['date'] < '2020-10-01'),"wave"] = 2
df.loc[(df['date'] > '2020-09-31') & (df['date'] < '2021-04-01'),"wave"] = 3
df.loc[(df['date'] > '2021-03-31') & (df['date'] < '2021-08-01'),"wave"] = 4
df.loc[(df['date'] > '2021-07-31') & (df['date'] < '2021-12-01'),"wave"] = 5
df.loc[(df['date'] > '2021-11-31'),"wave"] = 6
```

```python
df
```

```python
dff = pd.get_dummies(df, columns=['wave'])
dff
```

```python
dff["new_death/new_cases"]= dff['new_deaths']/dff['new_cases']
dff.replace([np.inf, -np.inf], 0, inplace=True)
dff.replace(np.nan, 0, inplace=True)
dff.replace(to_replace="NULL",value=0)
```

```python
dff = dff.groupby(by=["location","wave_1","wave_2","wave_3","wave_4","wave_5","wave_6"]).agg({'new_death/new_cases':'sum', 'population_density':'max'})
dff
```

```python
grouped_region_data = dff.reset_index()
grouped_region_data
```

```python
grouped_region_data.loc[grouped_region_data["wave_1"]==1,"wave_1"]=grouped_region_data["new_death/new_cases"]
grouped_region_data.loc[grouped_region_data["wave_2"]==1,"wave_2"]=grouped_region_data["new_death/new_cases"]
grouped_region_data.loc[grouped_region_data["wave_3"]==1,"wave_3"]=grouped_region_data["new_death/new_cases"]
grouped_region_data.loc[grouped_region_data["wave_4"]==1,"wave_4"]=grouped_region_data["new_death/new_cases"]
grouped_region_data.loc[grouped_region_data["wave_5"]==1,"wave_5"]=grouped_region_data["new_death/new_cases"]
grouped_region_data.loc[grouped_region_data["wave_6"]==1,"wave_6"]=grouped_region_data["new_death/new_cases"]

grouped_region_data.head(6)
```

```python
grouped_region_data = grouped_region_data.dropna()
grouped_region_data
```

```python
grouped_region_data = grouped_region_data.groupby(by=["location"]).agg({'wave_1':'sum','wave_2':'sum','wave_3':'sum','wave_4':'sum','wave_5':'sum','wave_6':'sum','population_density':'max'})
grouped_region_data
```

```python
grouped_region_data_final = grouped_region_data.reset_index()
grouped_region_data_final = grouped_region_data_final.sort_values(by=["population_density"], ascending=False)
```

```python
# grouped_region_data_final = grouped_region_data_final.drop([23])
grouped_region_data_final = grouped_region_data_final.drop(columns = ['population_density'])
```
